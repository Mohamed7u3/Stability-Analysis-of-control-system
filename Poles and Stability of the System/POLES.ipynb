# OVERVIEW
## ''' The poles are important when analyzing the stability of a system.'''
## We have 3 different Alternatives:
### 1. Asymptotically Stable System
### 2. Marginally Stable System
### 3. Unstable System


# Asymptotically Stable System.
## Each of the poles of the transfer function lies strictly in the left half plane (has strictly negative real part).

```python
import control
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
# Read Images
img = matplotlib.image.imread('Asymptotically Stable.png')
 
# Output Images
plt.imshow(img)
```

```python

# Define Transfer Function
num = np.array([2])
den = np.array([3 , 1])
H = control.tf(num , den)
print ('H(s) =', H)
# Poles
p = control.pole(H)
print ('p =', p)
control.pzmap(H)
# Step Response
t, y = control.step_response(H)
fig = plt.figure(figsize=(20, 4))
fig.add_subplot(1,2, 1)
plt.plot(t,y)
plt.title("Step Response")
plt.grid()
```

# Marginally Stable System.
## One or more poles lies on the imaginary axis (have real part equal to zero), and all these poles are distinct. Besides, no poles lie in the right half plane

```python
# Read Images
img = matplotlib.image.imread('Marginally Stable.png')
 
# Output Images
plt.imshow(img)
```

```python
# Define Transfer Function
num = np.array([3])
den = np.array([1, 0, 1])
H = control.tf(num , den)
print ('H(s) =', H)
# Poles
p = control.pole(H)
print ('p =', p)
control.pzmap(H)
# Step Response
tstart = 0; tstop = 20; tstep = 0.1
t = np.arange(tstart, tstop, tstep)
t, y = control.step_response(H, t)
fig = plt.figure(figsize=(20, 4))
fig.add_subplot(1,2, 1)
plt.plot(t,y)
plt.title("Step Response")
plt.grid()
```

# Unstable System.
## At least one pole lies in the right half plane (has real part greater than zero).

```python
# Read Images
img = matplotlib.image.imread('Unstable System.png')
 
# Output Images
plt.imshow(img)
```

```python
# Define Transfer Function
num = np.array([2])
den = np.array([3 , -1])
H = control.tf(num , den)
print ('H(s) =', H)
# Poles
p = control.pole(H)
print ('p =', p)
control.pzmap(H)
# Step Response
t, y = control.step_response(H)
fig = plt.figure(figsize=(20, 4))
fig.add_subplot(1,2, 1)
plt.plot(t,y)
plt.title("Step Response")
plt.grid()
```
